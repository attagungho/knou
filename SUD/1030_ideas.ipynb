```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize= (16, 8))

vectors = np.array([[2,1], [1,2]])
x = np.zeros((2,1))
y = np.zeros((2,1))

plt.subplot(1,2,1)
plt.quiver(x, y, vectors[:,0], vectors[:,1], scale=1, scale_units='xy', angles='xy', color=['m', 'y'])

plt.xticks(np.arange(-1, vectors[:,0].max()+3))  
plt.yticks(np.arange(-1, vectors[:,1].max()+3))
plt.grid()
plt.axvline(x=0, color='r', linestyle='dotted')
plt.axhline(y=0, color='r', linestyle='dotted')

plt.subplot(1,2,2)
adding_vectors = vectors[0] + vectors[1]
vectors = np.append(vectors, adding_vectors[None,:], axis=0)
x = np.zeros((3,1))
y = np.zeros((3,1))

plt.quiver(x, y, vectors[:,0], vectors[:,1], scale=1, scale_units='xy', angles='xy', color=['m', 'y', 'k'])

plt.xticks(np.arange(-1, vectors[:,0].max()+3))  
plt.yticks(np.arange(-1, vectors[:,1].max()+3))
plt.grid()
plt.axvline(x=0, color='r', linestyle='dotted')
plt.axhline(y=0, color='r', linestyle='dotted')

plt.show()
```
